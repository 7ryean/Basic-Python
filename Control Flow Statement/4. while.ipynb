# while

for문처럼 반복을 수행하는 문법

## while문 만들기

while문도 for문처럼 뒤에 콜론 ```:```을 붙여주고, 실행문은 꼭 들여쓰기를 해주어야 함
```python
while 조건문 :
    실행문
```

````python
a = 0

while a < 10 :
    print(a)
    a += 1
````

````python
account = 0

while account < 1000 :
    account += 100
    print(f'통장에 {account}만원을 적금했습니다.')
    if account == 1000 :
        print('통장 잔고가 천만원이 되었습니다.')
````

## 영원히 반복되는 while문

조건이 ```True```일 때 ```while```문은 계속해서 실행 됨. 조건문에 직접 ```True```값을 넣어주면? 반대로 ```False```값을 넣어주면?

````python
while True :
    print(1)
````

````python
while False :
    print(1)
````

## 특정 횟수만큼 반복되는 while문

```python
while 조건문 :
    실행문
    변화식
```

````python
cnt = 0

while cnt < 2 :
    print('삐까')
    cnt += 1
````

````python
cnt = 0

while cnt < 10 :
    print('삐까')
    cnt += 1
````

## for문을 while문으로 바꾸기

````python
for i in range(5) :
    print(i)
````

````python
i = 0

while i < 5 :
    print(i)
    i += 1
````

## 입력을 받아서 답을 맞추면 반복 종료되는 while문

정답을 맞출 때까지 퀴즈를 내는 while문을 만들고 싶다면? ```input```함수를 사용해 입력을 받고, 입력한 값이 정답인 경우 반복문을 종료

````python
while True :
    answer = input('꼬부기가 진화하면?')
    if answer == '어니부기' :
        print('정답!')
        break
    else :
        print('땡! 다시 한 번 시도해보세요')
````
